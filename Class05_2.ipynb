```python
street_names = elevator_accidents['House Street'].tolist()
house_numbers = elevator_accidents['House Number'].tolist()
house_numbers
```

```python
bad_places = [ num + "" + street for num,street in zip(house_numbers,street_names)]
print(bad_places)
```

```python
from collections import Counter
worst_places = Counter(bad_places).most_common(10)
print(worst_places)
```

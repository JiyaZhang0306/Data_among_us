```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'CTA_Data.csv'
ctas = pd.read_csv(path)
```

```python
plant_counts = ctas.Plants.value_counts()
display(plant_counts)
```

```python
plant_counts.plot(kind='bar', color='green')
plt.xlabel('Plant')
plt.ylabel('Number')
plt.title('Which kind of plant you want to be in a forest?')
plt.xticks(rotation=45)
```

```python
ctas.Hair_colors
```

```python
hair_color_counts = ctas.Hair_colors.value_counts()
display(hair_color_counts)
```

```python
# for plants and hair color
plants_and_hair = ctas.groupby(['Plants','Hair_colors']).size()
display(plants_and_hair)
```

```python
#Group by 'Plants' and 'Hair_colors' to get the counts
plants_and_hair = ctas.groupby(['Plants', 'Hair_colors']).size()

#put'Hair_colors' in one colums
plants_and_hair_unstacked = plants_and_hair.unstack(fill_value=0) 

# Plot the grouped data as a stacked bar chart
plants_and_hair_unstacked.plot(kind='bar')
plt.legend(bbox_to_anchor=(1.5, 1), loc='upper right')
plt.xticks(rotation=45)

# another one: make color together
plants_and_hair_unstacked.plot(kind='bar', stacked=True)
plt.legend(bbox_to_anchor=(1.5, 1), loc='upper right')

#colors = ['#FFFFFF', '#FFEF00', '#99FF99', '#67670B', '#0D57EF', '#B1710A', '#11D638', '#FBB6C6', '#FF0F00']

plt.title("Plant's haircolor")
plt.xlabel("Plant")
plt.ylabel("Number")

plt.xticks(rotation=45)
```

```python
#classfy dreams
ctas.Dreams
```

```python
dream_counts = ctas.Dreams.value_counts()
display(dream_counts)
```

```python
#count positive dreams -- including Unbelievable fantasies & Flying in the sky
df_dreams = pd.DataFrame(ctas, columns=['Dreams'])
positive_dreams = df_dreams[df_dreams['Dreams'].str.contains('Happy dreams')]
positive_dreams_count = positive_dreams.value_counts()
positive_dreams_count = positive_dreams.shape[0]
display(positive_dreams)
display(positive_dreams_count)
```

```python
#count negative dreams -- including Always forget my dream & rarely have dreamsncluding Stressful dreams & murder scene 
df_dreams = pd.DataFrame(ctas, columns=['Dreams'])
stressful_dreams = df_dreams[df_dreams['Dreams'].str.contains('Stressful dreams')]
murder_dreams = df_dreams[df_dreams['Dreams'].str.contains('murder scene')]
#connect stressful_dreams and murder_dreams
negative_dreams = pd.concat([stressful_dreams, murder_dreams])
negative_dreams_count = negative_dreams.shape[0]

display(negative_dreams)
display(negative_dreams_count)
```

```python
#count who don't have dreams -- including Always forget my dream & rarely have dreams
df_dreams = pd.DataFrame(ctas, columns=['Dreams'])
no_dream = df_dreams[df_dreams['Dreams'].str.contains('Always forget my dream')]
no_dream2 = df_dreams[df_dreams['Dreams'].str.contains('rarely have dreams')]
#connect stressful_dreams and rarely have dreams
no_dream = pd.concat([no_dream, murder_dreams])
no_dream_count = no_dream.shape[0]

display(no_dream)
display(no_dream_count)
```

```python
#make a new dictionary
dream_counts = pd.DataFrame({
    'Dream Type': ['Positive Dreams', 'Negative Dreams', 'No Dreams'],
    'Count': [positive_dreams_count, negative_dreams_count, no_dream_count]
})

# Plotting using .plot(kind='bar')
dream_counts.set_index('Dream Type').plot(kind='bar')

plt.title('CTAs Dream Category')
plt.xlabel('Dream Category')
plt.ylabel('Number')

plt.xticks(rotation=45)
```
